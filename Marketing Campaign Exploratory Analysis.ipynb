### Reading the file

```python
import pandas as pd
import numpy as np
df = pd.read_csv('campaign.csv', index_col=0)
```

*Show all columns*

```python
pd.set_option('display.max_columns', None) ## for easier navigating, i showed all the column names to see redundancies or unwanted columns
df
```

### Converting time stamp to date

```python
df=df.reset_index()
df['timestamp']= pd.to_datetime(df['timestamp'], unit ='s')
df
```

### Dropping columns with zero values

```python
df.describe()
```

```python
df = df.loc[:, (df != 0).any(axis=0)]
##(df!=0) create a boolean DataFrame where df is nonzero, 
##(.any(axis=0) return a boolean series showing which columns have nonzero values -- there will be one boolean value for each column
## df.loc is used to select those columns
## then reasssign df to drop all the zero-value columns
df ##25 columns were dropped now 110
```

#### DROP ALL COLUMNS WITH ID IN THE COLUMN NAME

```python
df = df[df.columns.drop(list(df.filter(regex='Id')))]
df ## 5 columns dropped -- now 105
# with data onhand, cant really use the ID columns to extract further info, will drop it
```

### DROP REDUNDANT COLUMNS

```python
df = df.drop(['last24hClicksOnAdvertiser', 'last24hClicksOnContainer', 'last24hClicksOnFormat'], axis=1) ## retain last 24h clicks
df = df.drop(['last24hImpressionsOnAdvertiser', 'last24hImpressionsOnContainer', 'last24hImpressionsOnFormat'], axis=1) ## retain last 24h impressions
df = df.drop(['last24hAdClosingsOnAdvertiser', 'last24hAdClosingsOnContainer', 'last24hAdClosingsOnFormat'], axis=1) ## retain last 24h closings
df = df.drop(['last30minClicksOnAdvertiser', 'last30minClicksOnContainer', 'last30minClicksOnFormat'], axis=1) ## retain last 30mn clicks
df = df.drop(['last30minImpressionsOnAdvertiser', 'last30minImpressionsOnContainer', 'last30minImpressionsOnFormat'], axis=1) ## retain last 30mn impressions
df = df.drop(['last30minAdClosingsOnAdvertiser', 'last30minAdClosingsOnContainer', 'last30minAdClosingsOnFormat'], axis=1) ## retain last 30mn closings
df = df.drop(['last7dClicksOnAdvertiser', 'last7dClicksOnContainer', 'last7dClicksOnFormat'], axis=1) ## retain last 7d clicks
df = df.drop(['last7dImpressionsOnAdvertiser', 'last7dImpressionsOnContainer', 'last7dImpressionsOnFormat'], axis=1) ## retain last 7d impressions
df = df.drop(['last7dAdClosingsOnAdvertiser', 'last7dAdClosingsOnContainer', 'last7dAdClosingsOnFormat'], axis=1) ## retain last 7d closings
df = df.drop(['last30dClicksOnAdvertiser', 'last30dClicksOnContainer', 'last30dClicksOnFormat'], axis=1) ## retain last 30d clicks
df = df.drop(['last30dImpressionsOnAdvertiser', 'last30dImpressionsOnContainer', 'last30dImpressionsOnFormat'], axis=1) ## retain last 30d impressions
df = df.drop(['last30dAdClosingsOnAdvertiser', 'last30dAdClosingsOnContainer', 'last30dAdClosingsOnFormat'], axis=1) ## retain last 30d closings
```

```python
df ## we're now left with 69 columns
```

If the median of the browser window width is 800 pixels, it means that half of the observations in the data set have a browser window width of 800 pixels or lower, and half have a width of 800 pixels or higher.

#### CLICK DURATION VS. CONVERSION

*Pearson's r, is a measure of the linear correlation between two variables. It ranges from -1 to 1 and tells you how closely two variables are related.*

```python
click_duration = df[['last30dClicks', 'last7dClicks','last24hClicks', 'last30minClicks','conversion']].copy()
click_duration.corr(method ='pearson')
```

```python
last30dClick = df['last30dClicks'].corr(df['conversion'])
last7dClick = df['last7dClicks'].corr(df['conversion'])
last24hrClick = df['last24hClicks'].corr(df['conversion'])
last30mnClick = df['last30minClicks'].corr(df['conversion'])
print("Correlation of converted customers who clicked in the last 30 days is:", last30dClick)
print("Correlation of converted customers who clicked in the last 7 days is:", last7dClick)
print("Correlation of converted customers who clicked in the last 24 hours is:", last24hrClick)
print("Correlation of converted customers who clicked in the last 30 mins. is:", last30mnClick)
```

*Those who have visited in the last 7 days or last 30 mins. have the higher chances of conversion.*

```python
date_click = df[['timestamp', 'click']].copy()
date_click = date_click[(date_click != 0).all(1)] ## remove all non clicks
date_click["timestamp"] = date_click["timestamp"].dt.round("H").dt.hour ### round off to hour
dc1 = date_click.groupby(['timestamp']).size().reset_index(name = 'Count')
```

```python
import matplotlib.pyplot as plt
plt.plot(dc1['timestamp'], dc1['Count'], color='blue', marker='o')
plt.title('CLICKS PER HOUR', fontsize=14)
plt.xlabel('time of the day', fontsize=14)
plt.ylabel('clicks', fontsize=14)
plt.grid(True)
plt.show()
```

*Most of the clicks happen at 10PM. Customers often visit around early morning (likely before going to work) and at night (after work)*

#### CONVERSION PER HOUR

```python
date_converted = df[['timestamp', 'conversion']].copy()
date_converted = date_converted[(date_converted != 0).all(1)] ## remove all non converted
date_converted["timestamp"] = date_converted["timestamp"].dt.round("H").dt.hour ### round off to hour
dc = date_converted.groupby(['timestamp']).size().reset_index(name = 'Count')
```

```python
plt.plot(dc['timestamp'], dc['Count'], color='green', marker='o')
plt.title('CONVERSION PER HOUR', fontsize=14)
plt.xlabel('time of the day', fontsize=14)
plt.ylabel('converted sales', fontsize=14)
plt.grid(True)
plt.show()
```

*Most conversion happen between 6 to 9am with 7 having the highest conversion of 36*
